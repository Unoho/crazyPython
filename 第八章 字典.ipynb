### 8.2.2 创建只包含一个元素的字典

```python
# 方法一：使用花括号创建字典
pennant_number = {16:'辽宁舰'}
print(pennant_number)
```

```python
# 方法二：使用dict()函数创建字典
pennant_number = dict(name='辽宁舰')
print(pennant_number)
```

### 8.2.3 创建字典的其他方法

```python
# 方法一：使用dict()函数和二元组列表创建包含两个元素的字典
pennant_numbers = dict([(16,'辽宁舰'),(17,'山东舰')])
print(pennant_numbers)
```

```python
# 方法二：使用zip()函数和dict()函数创建包含两个元素的字典
pennant_numbers = dict(zip((16,17),('辽宁舰','山东舰')))
print(pennant_numbers)
```

zip函数将(16,17),('辽宁舰','山东舰')按照顺序将舷号和船名打包为两个元组，再由dict函数转换为字典。

如下图所示，利用list函数将zip打包后的结果转换为列表，再打印出来。

```python
zipped = list(zip((16,17),('辽宁舰','山东舰'))) 
zipped
```

```python
# 方法三 使用dict.fromkeys方法
D = dict.fromkeys(['name','age','job'])
print(D)
```
